```python
driver=webdriver.Chrome()
```

```python
driver.get(r"https://us.puma.com/us/en/search?q=shoes&offset=96")
```

```python
titles=[]
category=[]
style_name=[]
price=[]
colors=[]
review=[]
star=[]
```

```python
star_tag=driver.find_elements(By.XPATH,'//div[@class="flex absolute overflow-hidden text-puma-black top-0 w-[94%]"]')
for i in star_tag:
    st=i.text
    star.append(st)
star
    
```

```python
titles_tag=driver.find_elements(By.XPATH,'//h3[@class="w-full mobile:text-sm mobile:pr-0 font-bold text-base pr-5 line-clamp-2"]')
for i in titles_tag[0:87]:
    title=i.text
    titles.append(title.split('\n')[0])
```

```python
titles
```

```python
price_tag=driver.find_elements(By.XPATH,'//div[@class="flex flex-col flex-none mobile:items-start items-end text-sm md:text-base mobile:mt-2"]')
for i in price_tag[0:87]:
    pricee=i.text
    price.append(pricee.split('\n')[0])
price
```

```python
style_tag=driver.find_elements(By.XPATH,'//h3[@class="w-full mobile:text-sm mobile:pr-0 font-bold text-base pr-5 line-clamp-2"]')
for i in style_tag[0:87]:
    style=i.text
    style_name.append(style.split('\n')[1])
```

```python
review_tag=driver.find_elements(By.XPATH,'//span[@class="ml-1 pl-px text-sm"]')
for i in review_tag[0:87]:
    reviw=i.text
    review.append(reviw.split('\n')[0])
review
```
